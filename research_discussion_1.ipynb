## Commercial Recommender System: Netflix

Netflix mainly uses collaborative filtering and content-based filtering. What collaborative filtering does is it recommends shows and movies based on what similar users watch and like. Content-based filtering looks at things like genre, actors, directors, and themes and recommends similar content.

I feel like, Netflix usually provides good recommendations. I've noticed that the more I watch, the better the recommendations become. Sometimes the system will get stuck recommending very similar films and shows. For example if I watch a documentary series or two, my recommendations may become filled with documentaries even when I want more variety.

## Attacks on Recommender Systems

The Washington Post article describes how IMDb users coordinated to negatively rate a movie before it was released. This is an example of an attack on a recommender system because users intentionally manipulated ratings to influence what the public thought.

A similar example is review bombing on websites such as IMDb, Rotten Tomatoes, and Steam. Groups of users sometimes leave large numbers of positive or negative reviews to influence ratings. Their reasons are typically unrelated to the quality of the movie or game.

To reduce this type of abuse I would require verified accounts and limit the influence of ratings from new users. I would also place less weight on ratings submitted before a movie is actually released and use algorithms to detect unusual voting patterns. These steps would at least make it more difficult for organized groups to manipulate ratings and recommendations.
